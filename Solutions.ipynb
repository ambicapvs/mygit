```python
#import required libraries
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import sklearn
from sklearn.ensemble import IsolationForest  
import plotly.express as px
```

```python
#load data
df = pd.read_csv("microai2.csv")
df.head(5)
```

```python
#training my model
mymodel = IsolationForest(n_estimators=50, max_samples='auto', contamination=float(0.1),max_features=1)
x = df[["Height"]]
mymodel.fit(x.values)
```

```python
#findind outliers
#x = df[["Height"]]
df['scores']=mymodel.decision_function(df[["Height"]])
df['anomaly']=mymodel.predict(df[["Height"]])
df.head(30)
```

```python
#outlier/anomaly list
anomaly=df.loc[df['anomaly']==-1]
anomaly_index=list(anomaly.index)
print(anomaly)
```
